# Result without imbalanced data

```python
df_without_method= {
    'accuracy': [0.9999950410165783, 0.9999915697281833, 0.9999950410165783, 0.9999950410165783],
    'balanced_accuracy': [0.5, 0.49999826434719535, 0.5, 0.5],
    'macro_precision': [0.49999752050828916, 0.4999975204996821, 0.49999752050828916, 0.49999752050828916],
    'macro_recall': [0.5, 0.49999826434719535, 0.5, 0.5],
    'macro_f1': [0.49999876025107065, 0.49999789242316206, 0.49999876025107065, 0.49999876025107065],
    'time_to_train': 1/np.array ([71.00356984138489, 272.9703857898712, 41.757519483566284, 1151.973237991333]),
    'time_to_test': 1/np.array([0.16847920417785645, 0.3677794933319092, 0.10965180397033691, 5.395903587341309])
    }
```

```python
df = pd.DataFrame(df_without_method, index = ['LR_original', 'DT_original', 'SGD_original','MLP_original']).T
df
```

```python
# set width of bar
barWidth = 0.15
fig = plt.subplots(figsize =(12, 8))

# set height of bar
a = df['LR_original']
b = df['DT_original']
c = df['SGD_original']
d = df['MLP_original']
# Set position of bar on X axis
br1 = np.arange(len(a))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]

# Make the plot
plt.bar(br1, a, color ='r', width = barWidth,
        edgecolor ='grey', label ='LR_original')
plt.bar(br2, b, color ='g', width = barWidth,
        edgecolor ='grey', label ='DT_original')
plt.bar(br3, c, color ='b', width = barWidth,
        edgecolor ='grey', label ='SGD_original')
plt.bar(br4, d, color ='y', width = barWidth,
        edgecolor ='grey', label ='MLP_original')

# Adding Xticks
plt.title('Result without imbalanced data')
plt.xticks([r + barWidth for r in range(len(a))],
        ['accuracy', 'balanced_accuracy', 'macro_precision', 'macro_recall',
       'macro_f1', '1/time_to_train', '1/time_to_test'])
plt.yscale('log')
plt.legend()
plt.show()
```
